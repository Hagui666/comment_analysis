*  將星級大於四星以及評論大於500責的店家篩選出來

```python
def select_data():
    df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/googleSearchFood/商家網址連結(已清洗)備分.csv')
    df['評論'] = df['評論'].replace('沒有評論', '0')
    df['星級'] = df['星級'].replace('沒有評論', '0')
    df['評論'] = df['評論'].str.replace(',', '')
    df['星級'] = df[['星級']].astype(float)
    df['評論'] = df[['評論']].astype(int)
    df = df[(df["評論"] > 500) & (df["星級"] > 4.0)]
    df['回復'] = ''
    df['地址'] = ''
    df.to_csv('/content/drive/MyDrive/Colab Notebooks/data/googleSearchFood/信義區餐廳_篩選評論.csv', encoding='utf-8-sig', index=False)
select_data()    
```

```python
bad = []
for i in commenttext:
  if i!=[]:
    s=SnowNLP(i[0])
    s1=SnowNLP(i[1])
    s2=SnowNLP(i[2])
    if s.sentiments < 0.2:
      print(i[0])
      print('待改善評論',s.sentiments)
      bad.append(i[0])
    if s1.sentiments < 0.2: 
      print(i[1])
      print('待改善評論',(s1.sentiments))
      bad.append(i[1])
    if s2.sentiments < 0.2:
      print(i[2])
      print('待改善評論',(s2.sentiments))
      bad.append(i[2])
```
